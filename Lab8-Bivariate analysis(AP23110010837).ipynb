```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('vgsales.csv')
data.head()
```

# Task
Generate separate Python code cells to create a line plot of global sales over time and a scatter plot of two continuous variables from the dataset, including styling and labeling for both plots.

## Data aggregation for line plot

### Subtask:
Aggregate the global sales data by year to prepare for the line plot.


**Reasoning**:
Aggregate the global sales by year for the line plot.



```python
yearly_sales = data.groupby('Year')['Global_Sales'].sum().reset_index()
display(yearly_sales)
```

## Generate line plot

### Subtask:
Create a line plot showing the trend of global sales over the years.


**Reasoning**:
Generate a line plot of global sales over time using the aggregated yearly sales data.



```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(yearly_sales['Year'], yearly_sales['Global_Sales'])
ax.set_title('Global Video Game Sales Over Time')
ax.set_xlabel('Year')
ax.set_ylabel('Global Sales (in millions)')
plt.show()
```

## Styling and labeling

### Subtask:
Enhance the line plot with appropriate titles, labels, and potentially different styles for better readability.


**Reasoning**:
Enhance the line plot with more appropriate titles, labels, and different styles for better readability.



```python
fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(yearly_sales['Year'], yearly_sales['Global_Sales'], marker='o', linestyle='-', color='skyblue')
ax.set_title('Global Video Game Sales Trend Over the Years (1980-2020)', fontsize=16)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Global Sales (in millions USD)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Generate scatter plot

### Subtask:
Create a scatter plot to visualize the relationship between two continuous variables in the dataset.


**Reasoning**:
Create a scatter plot to visualize the relationship between 'NA_Sales' and 'EU_Sales', adding title and axis labels.

## Styling and labeling (scatter plot)

### Subtask:
Add titles, labels, and customize the scatter plot for clarity.


**Reasoning**:
Generate the scatter plot with specified size, transparency, title, labels, and add a grid for clarity and display the plot.

## Subplots

Subplots are useful for displaying multiple plots within a single figure, making it easier to compare different visualizations.

### Method 1: Using `plt.subplots()`

This is the most common and recommended method for creating subplots. It returns a figure object and an array of axes objects.

```python
# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2, figsize=(8, 5))

# Plot on the first subplot (left)
axes[0].plot(yearly_sales['Year'], yearly_sales['Global_Sales'])
axes[0].set_title('Global Sales Over Time')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Global Sales (in millions)')

# Plot on the second subplot (right)
axes[1].scatter(data['NA_Sales'], data['EU_Sales'], alpha=0.5)
axes[1].set_title('NA Sales vs EU Sales')
axes[1].set_xlabel('North American Sales (in millions)')
axes[1].set_ylabel('European Sales (in millions)')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

### Method 2: Using `fig.add_subplot()`

This method is useful if you need to add subplots to an existing figure one by one.

```python
# Create a figure
fig = plt.figure(figsize=(12, 5))

# Add the first subplot (1 row, 2 columns, 1st plot)
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(yearly_sales['Year'], yearly_sales['Global_Sales'])
ax1.set_title('Global Sales Over Time')
ax1.set_xlabel('Year')
ax1.set_ylabel('Global Sales (in millions)')

# Add the second subplot (1 row, 2 columns, 2nd plot)
ax2 = fig.add_subplot(1, 2, 2)
ax2.scatter(data['NA_Sales'], data['EU_Sales'], alpha=0.5)
ax2.set_title('NA Sales vs EU Sales')
ax2.set_xlabel('North American Sales (in millions)')
ax2.set_ylabel('European Sales (in millions)')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

## Categorical-Continuous Plots

### Boxplot

**Subtask:** Create a boxplot to visualize the distribution of a continuous variable ('Global_Sales') across different categories of a categorical variable ('Genre').

**Reasoning:**
Generate a boxplot to show the distribution of Global Sales for each Genre, which will help identify differences in sales performance across different game genres.

```python
plt.figure(figsize=(8, 8))
sns.boxplot(data=data, x='Genre', y='Global_Sales') # Removed palette
plt.title('Global Sales Distribution by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Global Sales (in millions USD)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Bar Plot

**Subtask:** Create a bar plot to show the average of a continuous variable ('Global_Sales') for each category of a categorical variable ('Genre').

**Reasoning:**
Generate a bar plot to display the average Global Sales for each Genre, providing a clear comparison of the typical sales performance across different game genres.

```python
plt.figure(figsize=(8, 8))
sns.barplot(data=data, x='Genre', y='Global_Sales', estimator=np.mean) # Removed palette
plt.title('Average Global Sales by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average Global Sales (in millions USD)', fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Categorical-Categorical Plots

### Dodged Countplot

**Subtask:** Create a dodged countplot to visualize the distribution of one categorical variable ('Genre') within categories of another ('Platform').

**Reasoning:**
Generate a dodged countplot to show the count of each genre within each platform, which helps in comparing genre popularity across different platforms.

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=data, x='Platform', hue='Genre', palette='viridis')
plt.title('Count of Games by Platform and Genre (Dodged)', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### Stacked Countplot

**Subtask:** Create a stacked countplot to visualize the distribution of one categorical variable ('Genre') within categories of another ('Platform'), showing the total count for each category of the first variable.

**Reasoning:**
Generate a stacked countplot to show the proportion of each genre within each platform, providing an overview of the genre composition for each platform.

```python
# To create a stacked countplot, we can first create a crosstab
cross_tab = pd.crosstab(data['Platform'], data['Genre'])

cross_tab.plot(kind='bar', stacked=True, figsize=(10, 8), colormap='viridis')
plt.title('Stacked Count of Games by Platform and Genre', fontsize=16)
plt.xlabel('Platform', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=90)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
